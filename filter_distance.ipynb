```python
df = open('my_sites', 'r')
filterdata = open('my_filteredsites', 'w')
```

```python
correct = set([-3, -2, -1, 0, 1,2,3])

for i in df:
    line = i.strip().split(' ')
    #print(line)
    my_coords = set(map(float, line[1:]))
    #print(my_coords)
    #newch = dic_chr[ch]
    print(len(correct.intersection(my_coords)))
    if len(correct.intersection(my_coords)) != 0:
        filterdata.write(
            line[0] + '\t' 
            + line[1]+ '\t' 
            + line[2]+ '\t'
            + line[3]+ '\t' 
            + line[4] + "\n")
    else:
        continue
        
df.close()
filterdata.close()
```
