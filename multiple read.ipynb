```python
import pickle
from collections import Counter
```

```python
with open("my_new_for2.txt", "r") as a_file:
    with open('new_file_dict.txt', 'wb') as my_new:
        dic = {}
        for line in a_file:
            stripped_line = line.strip().split()
            if len(stripped_line)>1:
                if stripped_line[0] in dic:
                    continue
                else:
                    dic[stripped_line[0]] = stripped_line[1]
        pickle.dump(dic, my_new)
```

```python
dic
```

```python
for i in dic:
    my_num = dic[i]
    my_num_n = my_num.split(":")
    try:
        dic[i] = int(my_num_n[2])
    except:
        dic[i] = 0
```

```python
a = Counter(dic.values())
```

```python
#print(a)
```

```python
new_di={}
for i in a:
    if i != 0:
        new_di[i] = a[i]
```

```python
new_di[1000]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(new_di, density=True, bins=50, color='#0504aa') 
plt.ylabel('count')
plt.xlabel('n map')
plt.title('RNA')
plt.xticks(range(0, 1001, 100))
plt.savefig('reads_rna.png')
```
