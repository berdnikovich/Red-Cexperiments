Для начала соберем общую статистику по всем ридам по отдельности, для этого напишем будет итерироваться по полученным файлам и искать ключи из словаря, для которого мы потом построим диаграмму

```python
NH = open("my_data.txt", 'r')
```

```python
dic = {}
for i in NH:
    find_num = i.strip().split("\t")[1]
    #print(find_num)
    if find_num in dic.keys():
        #print('hey')
        dic[find_num] = dic[find_num] + 1
    else:
        dic[find_num] = 1
```

```python
dic
```

```python
output_dict = {}
for key in dic:
    output_dict[int(key)] = dic[key]
```

```python
output_dict
```

```python
a_file = open("data.pkl", "wb")
pickle.dump(output_dict, a_file)
a_file.close()
```

```python
for k,v in rna_dict.items():
    for k2,v2 in di_hist_rna.items():
        fr = float(k2.split(',')[0][1:])
        to = float(k2.split(',')[1][1:-1])
        if k >= fr and k < to:
            di_hist_rna[k2] = di_hist_rna[k2] + v
```
